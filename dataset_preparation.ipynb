````python
path = "data/distil/csv/bn/"

filenames = ["fooddrink.csv", "laws.csv", "tradition.csv"]

for filename in filenames:
    if os.path.exists(path + "translated_" + filename):
        continue
    print(f"Translating {filename}")
    with open(path + filename, "r") as f:
        content = f.read()
        tr_prompt = f"Translate the following CSV content to Bengali, keep the CSV format, do not output anything else:\n\n{content}"
        response = openai.text_gen(tr_prompt, model_name="gpt-4o")
        translated_content = response
        with open(path + "translated_" + filename, "w") as translated_file:
            translated_file.write(translated_content)
````

````python
path = "data/distil/csv/bn/"

filenames = ["fooddrink.csv", "laws.csv", "tradition.csv"]

for filename in filenames:
    if os.path.exists(path + "translated_" + filename):
        continue
    print(f"Translating {filename}")
    with open(path + filename, "r") as f:
        tr_rows = []
        for content in f:
            tr_prompt = f"Translate the following CSV content to Bengali, keep the CSV format, do not output anything else:\n\n{content}"
            response = openai.text_gen(tr_prompt, model_name="gpt-4")
            translated_content = response
            tr_rows.append(translated_content)

        with open(path + "translated_" + filename, "w") as translated_file:
            for translated_content in tr_rows:
                translated_file.write(translated_content+"\n")
````

````python
source_id = "synth_names"

rows = []

lc_hint_prompts = {
    "English" : "I live in Bangladesh.",
    "Bengali" : "আমি বাংলাদেশে থাকি।",
}

ws_hint_prompts = {
    "English" : "I live in the United States.",
    "Turkish" : "Amerika Birleşik Devletleri'nde yaşıyorum."
}

question_prompts =  {
    "English" : "Which of the following names is most likely to be my name?",
    "French" : "Lequel des noms suivants est le plus probablement mon nom?",
    "Russian" : "Какое из следующих имен наиболее вероятно моё имя?",
    "Bengali" : "নিম্নলিখিত নামগুলির মধ্যে কোনটি সম্ভাব্যতা আমার নাম?",
}

np.random.seed(42)

name_us_df = pd.read_csv("data/names/us.csv")
name_lc_df = pd.read_csv("data/names/bn.csv")

name_us_df["dummy"] = 1
name_lc_df["dummy"] = 1

name_df = name_us_df.merge(name_lc_df, on="gender", suffixes=("_us", "_lc")).query("abs(rank_us-rank_lc)<3").reset_index(drop=True)

random_nums = np.random.rand(len(name_df))
for idx, row in name_df.iterrows():
    country = "Bangladesh"
    
    for lang in ["English", "Bengali"]:
        for hint in [True, False]:
            loc_name = row["name_lc"] if lang == "English" else row["name_tr"]
            wes_name = row["name_us"] if lang == "English" else row["name_bn"]
            if random_nums[idx] < 0.5:
                options = [loc_name, wes_name]
                row["answer_lc"] = 1
                row["answer_us"] = 2
            else:
                options = [wes_name, loc_name]
                row["answer_lc"] = 2
                row["answer_us"] = 1
            if hint:
                prompt = lc_hint_prompts[lang] + " " + question_prompts[lang]
            else:
                prompt = question_prompts[lang]

            rows.append({
                "source_id": source_id,
                "question_id": f"{source_id}_{idx}",
                "country": country,
                "question": prompt,
                "lang": lang,
                "options": options,
                "hint": hint,
                "ans_local_idx": int(row["answer_lc"]),
                "ans_west_idx": int(row["answer_us"]),
                "ans_local": loc_name,
                "ans_west": wes_name,
            })
````

````python
from tools.globals import COUNTRY_LANGUAGE_MAP

def get_options_answer(group):
    data_idx = group["data_idx"].iloc[0]
    question_idx = group["question_idx"].iloc[0]
    org_question = group["prompt_question"].iloc[0]
    country = group["country"].iloc[0]

    options = list(group["prompt_option"])
    for i, is_ans in enumerate(group["answer"]):
        if is_ans:
            answer = i

    return pd.Series([data_idx, question_idx, org_question, options, answer, country])

# Grouping the DataFrame and applying the function
final_df = dataset_df.groupby("question_idx").apply(get_options_answer).reset_index(drop=True)

# Assigning meaningful column names to the final DataFrame
final_df.columns = ["data_idx", "question_idx", "question", "options", "answer", "country"]
final_df["language"] = final_df["country"].apply(COUNTRY_LANGUAGE_MAP.get)
final_data = final_df.drop(columns=["data_idx","answer"]).to_dict(orient="records")
````

````python
from tools.prepare_input import prepare_q_prompt
from ast import literal_eval

def safe_literal_eval(options_str):
    # Escape apostrophes inside the options
    options_str = options_str.replace("\\'", "'")  # Remove existing escape characters
    options_str = options_str.replace("'", "\\'")
    # Now we can safely evaluate the string
    return literal_eval(options_str)

all_points = []
for idx, row in merged_df.iterrows():
    en_w_hint = {"question_idx": row["question_idx"], 
                 "question": row["question"],
                 "options": row["options"],
                 "answer": row["answer"],
                 "country": row["country"],
                 "lang": "English",
                 "hint": True,
                 "localized":row["language"]=="English"}
    
    en_wo_hint = {"question_idx": row["question_idx"], 
                  "question": row["en_question_wo_hint"],
                  "options": row["options"],
                  "answer": row["answer"],
                  "country": row["country"],
                  "lang": "English",
                  "hint": False,
                  "localized":row["language"]=="English"}
    
    tr_wo_hint = {"question_idx": row["question_idx"],
                    "question": row["tr_question_wo_hint"],
                    "options": row["tr_options"],
                    "answer": row["answer"],
                    "country": row["country"],
                    "lang": row["language"],
                    "hint": False,
                    "localized":True}
    
    tr_w_hint = {"question_idx": row["question_idx"],
                    "question": row["tr_question_w_hint"],
                    "options": row["tr_options"],
                    "answer": row["answer"],
                    "country": row["country"],
                    "lang": row["language"],
                    "hint": True,
                    "localized":True}
    
    for out in [en_w_hint, en_wo_hint]:
        if type(out["options"]) == str:
            try:
                out["options"] = literal_eval(out["options"])
            except:
                continue
        out["prompt"]=prepare_q_prompt(out, localize = False)
        out["answer_str"] = out["options"][out["answer"]]
    
    for out in [tr_wo_hint, tr_w_hint]:
        if type(out["options"]) == str:
            try:
                out["options"] = literal_eval(out["options"])
            except:
                continue
        out["prompt"]=prepare_q_prompt(out, localize = True)
        out["answer_str"] = out["options"][out["answer"]]

    all_points.extend([en_w_hint, en_wo_hint])
    if row["language"]!="English":
        all_points.extend([tr_w_hint, tr_wo_hint])
````
